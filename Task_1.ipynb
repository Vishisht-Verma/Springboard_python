## Importing Pandas and Numpy

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

## Importing Dataset into the data_frame

```python
data_frame = pd.read_csv("./sample_data/healthcare-dataset-stroke-data.csv")
```

```python
data_frame.head()
```

```python
# @title gender

data_frame.groupby('gender').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title work_type

data_frame.groupby('work_type').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title age

data_frame['age'].plot(kind='hist', bins=20, title='age')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
